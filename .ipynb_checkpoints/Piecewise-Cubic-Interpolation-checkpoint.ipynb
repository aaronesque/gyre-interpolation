# Interpolation

### Cubic Hermite basis functions

```python
def tt(x0,x1, x):
    h = x1-x0
    t = (x-x0)/h
    return(t)

def H30(t):
    H30 = 1 - 3*t**2 + 2*t**3
    return(H30)

def H31(t):
    H31 = t - 2*t**2 + t**3
    return(H31)

def H32(t):
    H32 = -t**2 + t**3
    return(H32)

def H33(t):
    H33 = 3*t**2 - 2*t**3
    return(H33)
```

### Invoke HSE

```python
def dP_dx_n(mn, rhon, xn):
    dP_dx = -mn*rhon/(xn**2)
    return(dP_dx)
```

```python
dP_n = dP_dx_n(m_n, rho_n, x_n)
dP_n[0] = -3.
print(dP_n)
```

### Invoke Mass Conservation

```python
def pressure_n(x0,x1, P0,P1, dP0,dP1):
    x = sym.Symbol('x')
    h = (x1-x0)
    tau = tt(x0,x1, x)
    
    P_n = P0*H30(tau) + P1*H33(tau) \
            + h*( dP0*H31(tau) + dP1*H32(tau) )
    return(P_n)
```

```python
def dpressure_n(x0,x1, P0,P1, dP0,dP1):
    x = sym.Symbol('x')
    h = (x1-x0)
    tau = tt(x0,x1, x)
    
    pressure = pressure_n(x0,x1, P0,P1, dP0,dP1)
    dpressure = sym.diff( pressure, x )
    return(dpressure)
```

Shamelessly stolen from Mund, Hallet, and Henart (1975)

```python
def mass_n(x0,x1, P0,P1, dP0,dP1, m0):
    x = sym.symbols('x')
    h = (x1-x0) #am I introducing error here??
    
    dpres = dpressure_n(x0,x1, P0,P1, dP0,dP1)
    m1 = sym.sqrt(m0 - sym.integrate( x**4*dpres , (x,x0,x) ))
    
    return(m1)

def density_n(x0,x1, P0,P1, dP0,dP1, m0):
    x = sym.symbols('x')
    h = (x1-x0)
    
    mass = mass_n(x0,x1, P0,P1, dP0,dP1, m0)**(1./2)
    dens = sym.diff( mass )/(4*np.pi*x**2)

    return(dens)
```

```python
def lambdify(symbolic, xrange):
    x = sym.Symbol('x')
    lambdified = sym.lambdify(x, symbolic, "numpy")
    return(lambdified(xrange))
```

```python
def tau(xa,xb, x):
    h = xb-xa
    
    tau = (x-xa)/h
    
    return(tau)


def pressure(xa,xb, Pa,Pb, dPa,dPb):
    x = sym.Symbol('x')
    t = tau(xa,xb, x)
    h = xb-xa
    
    Pab = Pa*H30(t) + h*dPa*H31(t) + h*dPb*H32(t) + Pb*H33(t)
    
    return(Pab)


def dpressure(xa,xb, Pa,Pb, dPa,dPb):
    x = sym.Symbol('x')
    
    Pab = pressure(xa,xb, Pa,Pb, dPa,dPb)
    dPab = sym.diff( Pab, x )
    
    return(dPab)


def m2_integral(xa,xb, Pa,Pb, dPa,dPb):
    x = sym.Symbol('x')
    Pab = pressure(xa,xb, Pa,Pb, dPa,dPb)
    dPab = dpressure(xa,xb, Pa,Pb, dPa,dPb)
        
    m2ab = -8*np.pi*( sym.integrate( dPab*x**4 , (x, xa, xb) ) )

    return(m2ab)
```

```python
def P_int(xa,xb, Pa,Pb, dPa,dPb, ma,mb, n):
    h = (xb-xa)
    k = np.arange(n).tolist()
    x_c = [xa + h*(i+1)/(n+1) for i in k]
    
    P_c = [sym.Symbol(f'P_c{i}') for i in k]
    dP_c = [sym.Symbol(f'dP_c{i}') for i in k]
    
    Pac = pressure_n(xa,x_c[0], Pa,P_c[0], dPa,dP_c[0])
    dPac = dpressure_n(xa,x_c[0], Pa,P_c[0], dPa,dP_c[0])
    
    Pcc, dPcc = [],[]
    for i in k[:-1]:
        Pcc += [pressure_n(x_c[i],x_c[i+1], P_c[i],P_c[i+1], dP_c[i],dP_c[i+1])]
        dPcc += [dpressure_n(x_c[i],x_c[i+1], P_c[i],P_c[i+1], dP_c[i],dP_c[i+1])]
        
    Pcb = pressure_n(x_c[-1],xb, P_c[-1],Pb, dP_c[-1],dPb)
    dPcb = dpressure_n(x_c[-1],xb, P_c[-1],Pb, dP_c[-1],dPb)
    
    m2_sum = sum([ sym.integrate(dPcc[i]*x**4 , (x, x_c[i], x_c[i+1])) for i in k[:-1]] )
    
    m2_dif = -8*np.pi*( sym.integrate(dPac*x**4 , (x, xa,x_c[0])) \
                + sum([ sym.integrate(dPcc[i]*x**4 , (x, x_c[i], x_c[i+1])) for i in k[:-1]] ) \
                + sym.integrate(dPcb*x**4 , (x, x_c[-1],xb)) )
    
    P_c[0] = sym.solve(m2_dif - (mb**2 - ma**2), P_c[0] )[0]
    
    return(P_c, dP_c)
```

```python
def C1(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0):
    tau = sym.Symbol('tau')
    h = t1-t0
    mass = mass_n(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0)
    C1 = ((t0 + h*tau)**3)/mass
    return(C1)
    
def U(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0):
    tau = sym.Symbol('tau')
    h = t1-t0
    mass = mass_n(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0)
    dens = density_n(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0)
    U = b*((t0+h*tau)**3)*(dens / mass)
    return(U)
```

```python
def Vg(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0):
    tau = sym.Symbol('tau')
    Gamma = 1
    pres = pressure_n(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1)
    mass = mass_n(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0)
    dens = density_n(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0)
    Vg = -(a/Gamma)*(mass*dens)/(pres*( t0+h*tau ))
    return(Vg)
        
def V2(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0):
    tau = sym.Symbol('tau')
    h = t1-t0
    Gamma = 1
    Vg_n = Vg(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0)
    V2 = Gamma * Vg_n /((t0 + h*tau)**2)
    return(V2)

def As(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0):
    tau = sym.Symbol('tau')
    h = t1-t0
    U_n = U(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0)
    Vg_n = Vg(t0,t1, P0,P1, dP0,dP1, ddP0,ddP1, m0)
    As = 3 - U_n - ( (t0 + h*tau)/U_n) * (h)*sym.diff( U_n, tau) - Vg_n
    return(As)
```

```python
fig, ax = plt.subplots(1,3, figsize=(9,2.8), dpi=200)
    
t = np.linspace(0.0,1,100)
tau = sym.Symbol('tau')

for i in range(0,n-1):
    h = x_n[i+1]-x_n[i]
    x = x_n[i]+t*h
    
    C1_n = C1(x_n[i],x_n[i+1], P_n[i],P_n[i+1], dP_n[i],dP_n[i+1], ddP_n[i],ddP_n[i+1], m_n[i])
    #C1_n = lambdify(C1_n, t)
    U_n = U(x_n[i],x_n[i+1], P_n[i],P_n[i+1], dP_n[i],dP_n[i+1], ddP_n[i],ddP_n[i+1], m_n[i])
    U_n = lambdify(U_n, t)
    
    dlnC1_dlnx = lambdify( sym.diff( C1_n ,tau)*((x_n[i]+h*tau)/C1_n) , t)
    three_U = 3 - U_n
                    
    ax[0].plot(x, dlnC1_dlnx)
    ax[1].plot(x, three_U, ls='--', lw=1, label=r'$3-U$')
    
ax[0].set_xlabel('$x$ [$r$/$R$]', labelpad=0)
ax[0].set_ylabel('$dln$ $C_1$/$dln$ $x$', labelpad=0)

ax[1].set_xlabel('$x$ [$r$/$R$]', labelpad=0)
ax[1].set_ylabel('$3-U$', labelpad=0)
#ax[1].legend()

ax[2].set_xlim([0,1])
ax[2].set_ylim([0,1])
ax[2].annotate('They are the same!', xy=(0.1, 0.21), xytext=(0.1, 0.21))
ax[2].grid(None)
ax[2].tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
ax[2].tick_params(axis='y', which='both', left=False, right=False, labelleft=False)

fig.tight_layout()
fig.savefig('{eta}_dlnC1_3-U.png')
fig.show()
```

I get it now!! The reason $U$ (and hence $3-U$) works, despite my $C_1$ function exhibiting regions of undefined behavior, is because $U\propto m^2$ and $C_1 \propto \sqrt{m^2}$. In other words, the results depend on the square root of $m(0)^2 + P*t^5$, a quantity that may not be positive if the input isn't a valid stellar model!

Since $m \in \mathbb{I}$ just means $m^2 <0$, $U$ is well-defined but $C_1$ isn't.
