```python
import numpy as np
import sympy as sym
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Definitions for pressure/mass/etc function evaluations

# Define Hermite basis functions

def H3 (n, t):
    
    if n == 0:
        return 1 - 3*t**2 + 2*t**3
    elif n == 1:
        return t - 2*t**2 + t**3
    elif n == 2:
        return -t**2 + t**3
    elif n == 3:
        return 3*t**2 - 2*t**3
    else:
        raise Exception('Invalid n')
        
        
# Define the interval mapping function

def tau_func (x_ab, x):
    
    return (x - x_ab[0])/(x_ab[1] - x_ab[0])


# Define the pressure interpolating function

def P_func (x_ab, P_ab, dP_ab):
    
    x_ = sym.Symbol('x')
    t_ = tau_func(x_ab, x_)
    h = x_ab[1] - x_ab[0]
    
    return P_ab[0]*H3(0, t_) + h*dP_ab[0]*H3(1, t_) + h*dP_ab[1]*H3(2, t_) + P_ab[1]*H3(3, t_)


# Define the pressure gradient interpolating function

def dP_func (x_ab, P_ab, dP_ab):
    
    x_ = sym.Symbol('x')
    P_ = P_func(x_ab, P_ab, dP_ab)

    return sym.diff(P_, x_)


# Define the mass-squared interpolating function

def m2_func (x_ab, P_ab, dP_ab):
    
    x_ = sym.Symbol('x')
    dP_ = dP_func(x_ab, P_ab, dP_ab)
        
    return -8*np.pi*sym.integrate(x_**4*dP_, x_)
```

```python
### Definitions for optimization

# Define i'th inequality constraint for pressure gradient

def ineq_dP (y, i):
    
    # Extract pressures and pressure gradients
    
    n = len(y)//2
    
    P = y[:n]
    dP = y[n:]
    
    # Set up constraint
    
    C = -dP[i]
    
    # Return it
    
    return C


# Define Jacobian of ineq_dP

def jac_ineq_dP (y, i):
    
    # Extract pressures and pressure gradients
    
    n = len(y)//2
    
    P = y[:n]
    dP = y[n:]
    
    # Set up constraint Jacobian
    
    dC = np.zeros(2*n)
    dC[n+i] = -1.
    
    # Return it
    
    return dC


# Define i'th inequality constraint for pressure change

def ineq_delP (y, i, left, x_AB, P_AB, dP_AB):
    
    # Extract pressures and pressure gradients
    
    n = len(y)//2
    
    P = y[:n]
    dP = y[n:]
    
    # Set up positions of interior points
    
    x = np.linspace(x_AB[0], x_AB[1], n+2)[1:-1]
    
    # Calculate mean pressure slope
    
    if left:
        if i > 0:
            delP = (P[i] - P[i-1])/(x[i] - x[i-1])
        else:
            delP = (P[i] - P_AB[0])/(x[i] - x_AB[0])
    else:    
        if i < n-1:
            delP = (P[i+1] - P[i])/(x[i+1] - x[i])
        else:
            delP = (P_AB[1] - P[i])/(x_AB[1] - x[i])
            
    # Set up the constraint
    
    C = dP[i] - 3*delP
        
    # Return it
    
    return C

    
# Define Jacobian of ineq_delP

def jac_ineq_delP (y, i, left, x_AB, P_AB, dP_AB):
    
    # Extract pressures and pressure gradients
    
    n = len(y)//2
    
    P = y[:n]
    dP = y[n:]
    
    # Set up positions of interior points
    
    x = np.linspace(x_AB[0], x_AB[1], n+2)[1:-1]
    
    # Set up constraint Jacobian
    
    dC = np.zeros(2*n)
    
    if left:
        if i > 0:
            dC[i-1] = 3/(x[i] - x[i-1])
            dC[i] = -3/(x[i] - x[i-1])
        else:
            dC[i] = -3/(x[i] - x_AB[0])
    else:    
        if i < n-1:
            dC[i] = 3/(x[i+1] - x[i])
            dC[i+1] = -3/(x[i+1] - x[i])
        else:
            dC[i] = 3/(x_AB[1] - x[i])
        
    dC[n+i] = 1.
    
    # Return it
    
    return dC


# Define the equality constraint for mass conservation

def eq_mass (y, x_AB, P_AB, dP_AB, m_AB):
    
    # Extract pressures and pressure gradients
    
    n = len(y)//2
    
    P = y[:n]
    dP = y[n:]
     
    # Set up positions of interior points
    
    x = np.linspace(x_AB[0], x_AB[1], n+2)[1:-1]
    
    # Calculate the mass-squared sum
    
    x_ = sym.Symbol('x')

    m2 = 0
    
    for i in range(n+1):
        
        if i == 0:
            x_a = x_AB[0]
            P_a = P_AB[0]
            dP_a = dP_AB[0]
        else:
            x_a = x[i-1]
            P_a = P[i-1]
            dP_a = dP[i-1]
            
        if i == n:
            x_b = x_AB[1]
            P_b = P_AB[1]
            dP_b = dP_AB[1]
        else:
            x_b = x[i]
            P_b = P[i]
            dP_b = dP[i]
            
        m2_ = m2_func([x_a,x_b], [P_a,P_b], [dP_a,dP_b])
        
        m2 += m2_.evalf(subs={x_:x_b}) - m2_.evalf(subs={x_:x_a})
        
    # Set up the equality constraint
    
    C = m2 - (m_AB[1]**2 - m_AB[0]**2)
    
    # Return it
    
    return C

# Define second derivative variance function (to be minimized)

def var_d2P (y, args):
    
    # Extract pressures and pressure gradients
    
    n = len(y)//2
    
    P = y[:n]
    dP = y[n:]
    
    # Unpack extra args
    
    x_ab = args[0]
    P_ab = args[1]
    dP_ab = args[2]
    
    # Set up positions of interior points
    
    x = np.linspace(x_ab[0], x_ab[1], n+2)[1:-1]
    
    # Evaluate mean second derivative
    
    mean_d2P = (dP_ab[1] - dP_ab[0])/(x_ab[1] - x_ab[0])
    
    # Evaluate second derivative variance
    
    var_d2P = 0.
    
    for i in range(n+1):
        
        # Evaluate d2P at the start and end of the subinterval
        
        if i == 0:
            x_0 = x_ab[0]
            P_0 = P_ab[0]
            dP_0 = dP_ab[0]
        else:
            x_0 = x[i-1]
            P_0 = P[i-1]
            dP_0 = dP[i-1]
            
        if i == n:
            x_1 = x_ab[1]
            P_1 = P_ab[1]
            dP_1 = dP_ab[1]
        else:
            x_1 = x[i]
            P_1 = P[i]
            dP_1 = dP[i]
            
        d2P_0 =  6*(P_1 - P_0)/(x_1 - x_0)**2 - 2*(dP_1 + 2*dP_0)/(x_1 - x_0)
        d2P_1 = -6*(P_1 - P_0)/(x_1 - x_0)**2 + 2*(2*dP_1 + dP_0)/(x_1 - x_0)
    
        # Add in the integrated contribution to the variance
        
        var_d2P += (d2P_0**2 + d2P_1*d2P_1 + d2P_1**2)*(x_1 - x_0)/3
            
    var_d2P /= (x_ab[1] - x_ab[0])
        
    # Return it
    
    return var_d2P


# Interval optimization routine for a fixed number (n) of interior points

def opt_inter_fix (n, x_AB, P_AB, dP_AB, m_AB):
    
    # Check that n is positive
    
    if n <= 0:
        raise Exception('n is not positive')

    # Check that the boundary pressures are monotonic
    
    if P_AB[1] > P_AB[0]:
        raise Exception('P_b is greater than P_a')

    # Check that the boundary slopes are monotonic
    
    if dP_AB[0] > 0:
        raise Exception('dP_a is greater than 0')
        
    if dP_AB[1] > 0:
        raise Exception('dP_b is greater than 0')
        
    # Set up positions of interior points
    
    x = np.linspace(x_AB[0], x_AB[1], n+2)[1:-1]
    
    # Set up initial guesses for pressure and pressure slopes at interior points
    
    P = np.linspace(P_AB[0], P_AB[1], n+2)[1:n+1]

    dP = np.empty(n)
    dP[:] = (P_AB[0] - P_AB[1])/(x_AB[1] - x_AB[0])
    
    # Set up constraints list
    
    const_list = []
    
    # dP at interior points is negative
    
    for i in range(n):
        const_list += [{'type': 'ineq',
                        'fun': ineq_dP,
                        'jac': jac_ineq_dP,
                        'args': [i]}]
        
    # Pressure in subintervals is monotonic
    
    for i in range(n):
        const_list += [{'type': 'ineq',
                        'fun': ineq_delP,
                        'jac': jac_ineq_delP,
                        'args': [i, True, x_AB, P_AB, dP_AB]}]
        const_list += [{'type': 'ineq',
                        'fun': ineq_delP,
                        'jac': jac_ineq_delP,
                        'args': [i, False, x_AB, P_AB, dP_AB]}]
        
    # Mass is conserved
    
    const_list += [{'type': 'eq',
                    'fun': eq_mass,
                    'args': [x_AB, P_AB, dP_AB, m_AB]}]
    
    # Set up the solution vector
    
    y = np.append(P, dP)
    
    # Perform the optimization
    
    res = opt.minimize(var_d2P, y, [x_AB, P_AB, dP_AB], method='SLSQP', constraints=const_list,
                      options={'eps': 1e-6, 'ftol': 1e-12, 'maxiter': 1000, 'disp': True})
    
    # Set up results data
    
    if res.success:
        
        x = np.linspace(x_AB[0], x_AB[1], n+2)
    
        P = np.empty(n+2)
        P[0] = P_AB[0]
        P[1:n+1] = res.x[:n]
        P[n+1] = P_AB[1]
    
        dP = np.empty(n+2)
        dP[0] = dP_AB[0]
        dP[1:n+1] = res.x[n:]
        dP[n+1] = dP_AB[1]
        
        m2 = np.empty(n+2)
        m2[0] = m_AB[0]
        
        x_ = sym.Symbol('x')
        
        for i in range(n+1):
            x_ab = [x[i],x[i+1]]
            P_ab = [P[i],P[i+1]]
            dP_ab = [dP[i],dP[i+1]]
            m2_ = m2_func(x_ab, P_ab, dP_ab)
            m2[i+1] = m2[i] + m2_.evalf(subs={x_:x_ab[1]}) - m2_.evalf(subs={x_:x_ab[0]})
        
    else:
        
        x = None
        P = None
        dP = None
        m2 = None
        
    return x, P, dP, m2


# Interval optimization routine for a dynamic number of interior points

def opt_inter_dyn (x_AB, P_AB, dP_AB, m_AB):
    
    # Loop over n until we find a solution, or an upper limit is reached
    
    n_upper = 20
    
    for n in range(1, n_upper+1):
        
        x, P, dP, m2 = opt_inter_fix(n, x_AB, P_AB, dP_AB, m_AB)
        
        # Check for completion
        
        if x is not None:
            return x, P, dP, m2
            
    # Return a null result
    
    return None, None, None, None


# Plot data

def plot_data (x, P, dP, m2, n_over=25):
    
    # Check on x, P and dP
    
    if x is None:
        raise Exception('x is None')
    if P is None:
        raise Exception('P is None')
    if dP is None:
        raise Exception('dP is None')
    if m2 is None:
        raise Exception('m2 is None')
        
    # Set up interpolation functions
    
    n = len(P)-2
    
    x_ = sym.Symbol('x')
    
    P_f = []
    dP_f = []
    m2_f = []
    
    for j in range(n+1):

        x_ab = [x[j],x[j+1]]
        P_ab = [P[j],P[j+1]]
        dP_ab = [dP[j],dP[j+1]]
        
        P_f += [sym.lambdify(x_, P_func(x_ab, P_ab, dP_ab), 'numpy')]
        dP_f += [sym.lambdify(x_, dP_func(x_ab, P_ab, dP_ab), 'numpy')]
        m2_f += [sym.lambdify(x_, m2_func(x_ab, P_ab, dP_ab), 'numpy')]
    
    # Set up the interpolation x
    
    xi = np.linspace(x[0], x[-1], (n+1)*n_over+1)
    
    ni = len(xi)
    
    # Calculate the interpolated pressure, pressure gradient and mass
    
    Pi = np.empty(ni)
    dPi = np.empty(ni)
    m2i = np.empty(ni)
    
    for i in range(ni):

        j = np.where(xi[i] >= x[:n+1])[0][-1]
        
        Pi[i] = P_f[j](xi[i])
        dPi[i] = dP_f[j](xi[i])
        m2i[i] = m2[j] + m2_f[j](xi[i]) - m2_f[j](x[j])

    # Do the plot
    
    fig, ax = plt.subplots(3, figsize=[6,18])
    
    ax[0].plot(xi, Pi)
    ax[0].scatter(x, P)
    ax[0].set_xlabel('x')
    ax[0].set_ylabel('P')
    
    ax[1].plot(xi, dPi)
    ax[1].scatter(x, dP)
    ax[1].axhline(y=0, ls='--', c='orange')
    ax[1].set_xlabel('x')
    ax[1].set_ylabel('dP/dx')

    ax[2].plot(xi, np.sqrt(m2i))
    ax[2].scatter(x, np.sqrt(m2))
    ax[2].set_xlabel('x')
    ax[2].set_ylabel('m')
    
```

```python
# Set up input state

x_AB = [0,1]
P_AB = [1,0]
dP_AB = [0,-0.2]
m_AB = [0,2]

n = 3

# Perform the optimization

x, P, dP, m2 = opt_inter_dyn( x_AB, P_AB, dP_AB, m_AB)  #opt_inter_fix(n, x_AB, P_AB, dP_AB, m_AB) #

plot_data(x, P, dP, m2)

```
